## Model

```python
x=car.drop(columns="Price")
y=car["Price"]
```

```python
x # this is our feature
```

```python
y # our output
```

```python
from sklearn.model_selection import train_test_split # is used to split a dataset into training and testing sets
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2) 

# the function return four errors X_train ,X_test,y_train,y_test after knowing this values we can use it in our model to evalute its performace on unseen data
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder() #OneHotEncoder is a preprocessing technique in machine learning used to transform categorical data into a numerical format that can be used in machine learning algorithms. 
ohe.fit(x[['name','company','fuel_type']])
```

```python
ohe.categories_ #stores information about the categories it has encountered during the fitting process. 
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder="passthrough")

#This transformer is used for applying different preprocessing steps to different columns in your dataset
```

```python
column_trans
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)

#This function is used to create a scikit-learn pipeline that sequentially applies a series of transformers and a final estimator (machine learning model). 
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
y_pred
```

```python
 y_test.shape
```

```python
y_pred.shape
```

```python
y_test
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
scores=[]
for i in range(10):
    X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    print(r2_score(y_test,y_pred),i)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
